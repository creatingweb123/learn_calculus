```python
def replace_words(words, search_word, update_word):
    
    words = words.split()
    for i,  word in enumerate(words):
        if search_word.lower() in word.lower():  # 'Yesterday,', 'yesterday'
            if word[0].isupper():
                # Return a copy with all occurrences of substring old replaced by new.
                word = word.replace(word[:-1], update_word)
                word = word.replace(word[0], update_word[0].upper())
            else:
                word = word.replace(word, update_word)
            words[i] = word
    return words
```

```python
replace_words('Mm mm mm mm mm mm mm', "mm", "oh")
```

```python
from pprint import pprint

words = lyrics.split('\n')

search_word = "yesterday"
update_word = "today"
lines = []

for i in range(len(words)-1):
    lines.append(replace_words(words[i], search_word, update_word))

#print(lines)
pprint(lines, width=100)
```

```python
with open("today.txt", 'wt', encoding="utf-8") as fp:
    for i in range(len(lines)-1):
        fp.write(" ".join(lines[i]))
        fp.write('\n')
```
